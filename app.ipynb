```python
import os
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray
from sklearn.model_selection import train_test_split
# import seaborn as sns
```

```python
from keras.models import Sequential, load_model
from keras import layers
from keras.layers import Conv2D, MaxPooling2D, Activation,Flatten, Dense
```

```python
data_path = "D:/potato/potato_des/potato_des/data/"
```

```python
dataset = os.listdir(data_path) 
```

```python
list_1 = []  # labels 
list_2 = []  # images convert into array
for i in dataset:
    list_1.append(int(i[0]))
    img = Image.open(data_path + i)
    img = img.resize((100,100))
    data = asarray(img)
    list_2.append(data)
```

```python
print(list_1)
```

```python
images = np.array(list_2)
labels = np.array(list_1)
```

```python
type(images)
type(labels)
```

```python
images.shape
```

```python
labels.shape
```

```python
labels
```

```python
train_images,test_images,train_labels,test_labels = train_test_split(images,labels,test_size=0.2,random_state=41)
```

```python
len(train_images)
```

```python
len(test_images)
```

```python
len(train_labels)
```

```python
len(test_labels)
```

```python
categories=["potato_early_blight","potato_healthy","potato_late_blight"]
```

```python
len(categories)
```

```python
train_labels
```

```python
train_images,test_images = train_images/255.0,test_images/255.0
```

```python
train_images
```

```python
model = Sequential()
model.add(Conv2D(64,(3,3),activation='relu',input_shape=(100,100,3)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D(2,2))
```

```python
model.summary()
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(30,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics = ['accuracy'])
# model.save('my_model.h5')
```

```python
flat = model.fit(train_images,train_labels,epochs=10,validation_data=(test_images,test_labels))
```

```python
plt.plot(flat.history['accuracy'],label ='accuracy')
plt.plot(flat.history['val_accuracy'],label ='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5,1])
plt.legend(loc='upper right')
```

```python
test_acc = model.evaluate(test_images,test_labels)
```

```python
print(test_acc)
```
